```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
diff_function = lambda t, s: np.exp(-t)
s0 = -1
t0 = 0
tf = 1 
h = 0.01
t = np.arange(t0, tf+h, h)
```

```python
def runge_kutta_2(f, t, s0):
    l = len(t)
    s = np.zeros(l)
    s[0] = s0
    h = t[1]-t[0]
    
    for i in range(l - 1):
        k1 = f (t[i], s[i])
        
        # k2 = f( t[i]+h, s[i]+h*k1 )
        argument1_k2 = t[i]+h
        argument2_k2 = s[i] + h*f(t[i], s[i])
        k2 = f(argument1_k2, argument2_k2)
        
        s[i+1] = s[i] + 0.5*(k1 + k2)*h
        
    return s
```

```python
RK2_app = runge_kutta_2 (diff_function, t, s0)
```

```python
plt.figure(figsize = (8,5))
plt.plot(t, RK2_app, "b.-", label = "Approximation")
plt.plot(t, -np.exp(-t), "g", label = "Exact")
plt.title("Approximate and Exact solution for simple ODE")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid()
plt.legend(loc = "lower right")
```
