```python
# Bouncing Ball trajectory
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.8
t0 = 0
tf = 10
dt = 0.01
t = np.arange(t0, tf+dt, dt)
x = np.zeros(len(t))
v = np.zeros(len(t))
x[0] = 1.0
v[0] = 0
```

```python
for i in range(1000):
    x[i+1] = x[i] + v[i]*dt
    v[i+1] = v[i] - g*dt
    
    if x[i+1]<0:
        x[i+1] = -x[i+1]
        v[i+1] = -v[i+1]
        
```

```python
plt.plot(t,x)
```
