```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# diff_function = lambda y,x : x*y
diff_function = lambda y,x: np.exp(-x)
```

```python
y0 = -1
h = 0.01
x0 = 0
xf = 1
x = np.arange(x0, xf+h, h)
```

```python
def RK4(f, y0, x):
    l = len(x)
    h = x[1] - x[0]
    
    y = np.zeros((l))
    y[0] = y0
    
    for i in range(l-1):
        
        k1 = f(y[i], x[i])
        k2 = f(y[i]+k1*h*0.5 , x[i]+0.5*h)
        k3 = f(y[i]+k2*0.5*h, x[i]+0.5*h)
        k4 = f(y[i]+k3*0.5*h, x[i]+h)
        
        y[i+1] = y[i] + h*(k1+ 2*k2 + 2*k3 + k4)/6
        
    return y
```

```python
RK4_app = RK4(diff_function, y0, x)
```

```python
plt.figure(figsize = (8,5))
plt.plot(x, RK4_app, "b.-", label = "Approximation")
plt.plot(x, -np.exp(-x), "g", label = "Exact")
plt.title("Approximate and Exact solution for simple ODE")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid()
plt.legend(loc = "lower right")
```
